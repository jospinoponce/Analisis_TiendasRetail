# Data Source

```python
# Leer los CSV 

tipos_tiendas_totales= {
    'CD'                                          : 'str',    
    'Año'                                         : 'str'
}

tipos_mismas_tiendas= {
    'CD'                                          : 'str',    
    'Año'                                         : 'str'
}

tiendas_totales = pd.read_csv("tiendas_totales.csv", delimiter=";",  encoding="windows-1252", 
                              low_memory=False, dtype=tipos_tiendas_totales )

mismas_tiendas =  pd.read_csv("mismas_tiendas.csv",  delimiter=";",  encoding="windows-1252", 
                              low_memory=False, dtype=tipos_mismas_tiendas)
```

# 1 Productos

## 1.1 Unión Tiendas Totales Y Mismas Tiendas

fusiona los dos datasets.

```python
# Unir los DataFrames y agregar una columna indicador
merged_tiendas_totales= pd.merge(mismas_tiendas, tiendas_totales, how='left', indicator=True)


# Filtrar solo las filas de mismas_tiendas que están en tiendas_totales
result_df = merged_tiendas_totales[merged_tiendas_totales['_merge'] == 'left_only']

#Eliminar la columna indicador, renombrar 
result_df = result_df.drop(columns=['_merge'])
result_df['Tipo_tienda']="Mismas Tiendas"
result_df


#crear columna Tipo_Tienda en df
tiendas_totales['Tipo_tienda']='Tiendas Totales'


#concatenar las dos tablas
df_concat = pd.concat([tiendas_totales,result_df])



#Vista
df_concat

#Exportar el dataframe
#df_concat.to_csv("data_BI/productos.csv", index=False, sep=';', encoding='windows-1252')
```
